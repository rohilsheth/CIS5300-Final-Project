```python
import requests
```

```python
#group
def gender_check(json_var):
  #individual artist
  if len(json_var['members']) == 0:
    gender = json_var['gender']
    return gender
  male_count = 0
  female_count = 0
  for member in json_var['members']:
    if member['gender'] == 'Male':
      male_count += 1
    elif member['gender'] == 'Female':
      female_count += 1
  if male_count > 0 and female_count == 0:
    gender = 'Male'
  if female_count > 0 and male_count == 0:
    gender = 'Female'
  if male_count > 0 and female_count > 0:
    gender = 'Mixed'
  if male_count == 0 and female_count == 0:
    gender = 'Unknown'
  return gender

def song_count(json_var):
  count = 0
  for album in json_var['albums']:
    for song in album['songs']:
      count +=1
  return count
def lyrics_check(song):
  if len(song['lyrics']) == 0:
    return False
  if song['language_detect'] != 'english':
    return False
  if len(song['lyrics'].split()) < 10:
    return False
  if song['publicationDate'] == '':
    return False
  return True
```

```python
import pandas as pd
songs_df = pd.DataFrame(columns = ['Name','Song Title','Publication Date','Gender','Lyrics'])
songs_df
```

```python
songs = []
rejected_artists = 0
rejected_songs = 0
for i in range(0,77600,200):
  print('iteration: ',i)
  x = requests.get(f'https://wasabi.i3s.unice.fr/api/v1/artist_all/{i}')
  xj = x.json()
  for artist in xj:
    artist_name = artist['name']
    try:
      gender = gender_check(artist)
      gender_error = False
    except:
      gender_error = True
      rejected_artists +=1
    if song_count(artist) > 5 and not gender_error:
      for album in artist['albums']:
        for song in album['songs']:
          if lyrics_check(song):
            title = song['title']
            date = song['publicationDate']
            lyrics = song['lyrics']
            try:
              s = ''
              genres = song['genre']
              for genre in genres:
                s = s + genre + ','
              genre_str = s
            except:
              genre_str = 'N/A'
            songs.append([artist_name, title, date, gender, genre_str, lyrics])
    else:
      rejected_songs += 1
  print("list length: ",len(songs))
print('rejected_songs',rejected_songs)
print('rejected_artists',rejected_artists)
```

```python
import pandas as pd
df = pd.DataFrame(songs)
```

```python
df.columns = ['Name','Song Title','Publication Date','Gender','Genre','Lyrics']
```

```python
df2 = df.drop_duplicates()
df2
```

```python
df3 = df2.drop_duplicates(subset=['Lyrics'])
```

```python
df3 = df3.reset_index()
```

```python
df3.to_csv('530datasetV1.csv')
```

```python
from sklearn.model_selection import train_test_split
train, dev_test = train_test_split(df3, test_size = 0.2)
```

```python
dev, test = train_test_split(dev_test, test_size = 0.5)
```

```python
train.to_csv('530datasetV1_train.csv')
dev.to_csv('530datasetV1_dev.csv')
test.to_csv('530datasetV1_test.csv')

```
